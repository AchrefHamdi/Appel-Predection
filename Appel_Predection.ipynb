```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/AchrefHamdi/Appel-Predection/main/AAPL.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], high=data["High"],
                                        low=data["Low"], close=data["Close"])])
figure.update_layout(title = "Apple Stock Price", xaxis_rangeslider_visible=False)
figure.show()
```

```python
from autots import AutoTS
model = AutoTS(forecast_length=5, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
